```python
%matplotlib inline

import pandas as pd
```

```python
data_original = pd.read_csv("covid_threads - ANNOTATED.csv")
data_original.rename(columns={
    data_original.columns[10]: 'account_kind',
    data_original.columns[11]: 'thread_kind'},
    inplace = True)
```

```python
data_original
```

```python
data_clean = data_original.dropna(subset=['date'])
data_clean = data_clean[data_clean['thread_kind'] != 'REPEAT']
data_clean['thread_kind'] = data_clean['thread_kind'].str.strip()
```

```python
data_clean.shape
```

```python
data_clean
```

```python
data_clean['thread_kind'].value_counts().plot(kind='bar', rot=45)
```

```python
data_clean['account_kind'].value_counts().plot(kind='bar', rot=45)
```

```python
data_clean['locality of content'].value_counts().plot(kind='bar', rot=55, figsize=(15,3))
```

```python
order = ['summary of article', 'list of articles', 'summary of news', 'live tweeting', 'opinion', 'OTHER', 'tweetorial']
# data_clean.groupby(['thread_kind', 'account_kind']).size().unstack().plot(kind='bar', stacked=True, rot=55, figsize=(10,5))
gb = data_clean.groupby(['thread_kind', 'account_kind']).size().unstack().loc[order]
gb.plot(kind='bar', stacked=True, rot=30, figsize=(20,10), fontsize=20).legend(loc=1, prop={'size': 15})
```
